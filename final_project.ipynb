```python
import json

class InvertedIndex:
    def __init__(self, path):
        self.path = path
        self.index = {}

    def build_inverted_index(self):
        documents = {}

        with open(self.path, encoding='utf-8') as f:
            for line in f:
                doc_id, text = line.split("\t", 1)
                documents[int(doc_id)] = text

        index = {}

        for key, value in documents.items():
            cleaned = ''

            for ch in value:
                if ch.isalpha() or ch == ' ':
                    cleaned += ch.lower()

            words = set(cleaned.split())

            for word in words:
                if word not in index:
                    index[word] = {key}
                else:
                    index[word].add(key)

        self.index = index

    def query(self, words):
        result = None

        for word in words:
            docs = self.index.get(word, set())

            if result is None:
                result = docs
            else:
                result = result & docs

        return list(result) if result else []

    def dump(self, filepath):
        data = {k: list(v) for k, v in self.index.items()}

        with open(filepath, 'w', encoding='utf-8') as f:
            json.dump(data, f)

    @classmethod
    def load(cls, filepath):
        with open(filepath, 'r', encoding='utf-8') as f:
            data = json.load(f)

        obj = cls(None)
        obj.index = {k: set(v) for k, v in data.items()}

        return obj
```

```python
print(index.query(["flexible"]))
print(index.query(["animalia", "book"]))

index.dump("index.json")

new_index = InvertedIndex.load("index.json")

print(new_index.query(["animalia"]))
```
