```python
from typing import Any, Dict, List, Tuple


data_1 = {'a': 123, 'b': 23, 'c': 0}
data_2 = {'a': 1, 'b': 11, 'd': 99}
def task_1(data_1: Dict[str, int], data_2: Dict[str, int]):
    r = {}
    for i in data_1.keys():
        if i in data_2.keys():
            r[i] = data_1[i] + data_2[i]
        else:
            r[i] = data_1[i]
    for i in data_2.keys():
        if i not in data_1.keys():
            r[i] = data_2[i]
    return r
print(task_1(data_1, data_2))
```

```python
def task_3(data: Dict[Any, List[str]]):
    a = ['']

    for b in data.values():
        c = []

        for x in a:
            for y in b:
                c.append(x + y)

        a = c

    return a
data = {'1': ['a', 'b'], '2': ['c', 'd']}
print(task_3(data))
```

```python
data = {'a': -1}
def task_4(data: Dict[str, int]):
    if data == {}:
        return []
    elif len(data) < 3:
        return list(data.keys())
    else:
        r = []
        max_values = data.values()  # finding 3 max values
        max_values = sorted(max_values, reverse=True)[:3]
        for i in data.keys():
            if data[i] in max_values:
                r.append(i)
    return r
print(task_4(data))
```

```python
data = [('yellow', 1), ('blue', 2), ('yellow', 3), ('blue', 4), ('red', 1)]
def task_5(data: List[Tuple[Any, Any]]) -> Dict[str, List[int]]:
    d = {}
    for i in data:
        if i[0] in d.keys():
            d[i[0]].append(i[1])
        else:
            d[i[0]] = [i[1]]
    return d
print(task_5(data))
```

```python
def task_6(data: List[Any]):
    return list(set(data))
print(task_6([1, 1, 3, "3"]))
    
```
